###### A. Write a program

```python
def categorize(condition):
    if condition =='Completed':
        return 'Completed'
        
    elif condition == 'Defaulted' or condition == 'Chargedoff':
        return 'Defaulted'
    
Pri_load['LoanStatus'] = Pri_load.apply(lambda x: categorize(x['LoanStatus']) , axis=1)
```

    Sample output : 


    Debt Consolidation    106

```python
categories ={1: 'Debt Consolidation', 2: 'Home Improvement', 3: 'Business', 6: 'Auto', 7: 'Other'}
```

```python
def cat (a):
    if a in categories.keys():
        return(categories[a])
```

```python
Pri_load['ListingCategory (numeric)'] = Pri_load.apply(lambda x: cat(x['ListingCategory (numeric)']) , axis=1)
```
